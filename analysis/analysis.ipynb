```python
%matplotlib inline
from plotlib import plot_bundle_avgs, region_split, bundle_average, MongoCentral, get_mongo_db_client
from IPython import display
from IPython.core.display import display as dplay, HTML
import sys
dplay(HTML("<style>.container { width:100% !important; }</style>"))
root_dir = "/home/nick/Project/pymarl/src"
sys.path = [root_dir] + sys.path # adjust to local folder structure


mongo_central = MongoCentral(conf_names=["gandalf_pymarl"], 
                             root_dir=root_dir)
```

```python

#2s3z 12x12 qmix


data__conv_ddpg_input_grid = mongo_central.get_name_label_names("", "conv_ddpg_input_grid_2s3z_12x12_qmix_nick_test_2")
print(data__conv_ddpg_input_grid)

data__mathias = mongo_central.get_name_label_names("", "mathias_2s3z_12x12_qmix_nick_test_2")
print(data__mathias)

data__conv_ddpg = mongo_central.get_name_label_names("", "conv_ddpg_2s3z_12x12_qmix_nick_test_2")
print(data__conv_ddpg)

data__rnn = mongo_central.get_name_label_names("", "rnn_2s3z_12x12_qmix_nick_test_2")
print(data__rnn)

data__rnn_input_action = mongo_central.get_name_label_names("", "rnn_input_action_2s3z_12x12_qmix_nick_test_2")
print(data__rnn_input_action)


bavg = bundle_average(mongo_central, 
                      [("conv_input_flat_agent", data__mathias, "test_battle_won_mean", "test_battle_won_mean_T"),
                       ("rnn_input_agent", data__rnn_input_action, "test_battle_won_mean", "test_battle_won_mean_T"),
                       ("rnn_agent", data__rnn, "test_battle_won_mean", "test_battle_won_mean_T")
                      ], 
                      no_fails=True, 
                      window=20000)

bavg2 = bundle_average(mongo_central, 
                      [("conv_input_grid_agent", data__conv_ddpg_input_grid, "test_battle_won_mean", "test_battle_won_mean_T"),
                       ("conv_agent", data__conv_ddpg, "test_battle_won_mean", "test_battle_won_mean_T"),
                       ("rnn_agent", data__rnn, "test_battle_won_mean", "test_battle_won_mean_T")
                      ], 
                      no_fails=True, 
                      window=20000)

a = plot_bundle_avgs(bavg,
                     figsize=(15,8))

b = plot_bundle_avgs(bavg2,
                     figsize=(15,8))

```
